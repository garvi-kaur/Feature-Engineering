Feature Engineering

1. What is a parameter?

A parameter is a value that the model learns during training. For example, in linear regression, the slope and intercept are parameters that define the model.



2. What is correlation?

Correlation is a statistical measure that describes the strength and direction of a linear relationship between two variables. It ranges from -1 to 1.



3. What does negative correlation mean?

Negative correlation means that as one variable increases, the other decreases. A correlation value closer to -1 indicates a strong negative relationship.

5. How does loss value help in determining whether the model is good or not?


Loss value quantifies the error between predicted and actual values. A smaller loss indicates a better-performing model.

6. What are continuous and categorical variables?


Continuous variables: Numeric values that can take any value in a range (e.g., temperature, height).

Categorical variables: Variables with a fixed number of categories or labels (e.g., gender, color).

8. What do you mean by training and testing a dataset?


Training involves using data to teach the model. Testing evaluates the model's performance on unseen data to check generalization.

9. What is sklearn.preprocessing?

sklearn.preprocessing is a Scikit-learn module with functions to preprocess data like scaling, encoding, normalization, etc.

10. What is a Test set?


A test set is a portion of data used to assess the performance of a trained model. It contains data not seen during training.

17. What is causation? Explain difference between correlation and causation with an example.

Causation means one event directly causes another.
Correlation is just a relationship, not cause-effect.
Example: Ice cream sales and drowning deaths are correlated (due to summer) but one doesnt cause the other.

18. What is an Optimizer? What are different types of optimizers? Explain each with an example.

An optimizer updates model parameters to minimize the loss.
Types:

SGD: Stochastic Gradient Descent

Adam: Combines momentum and RMSprop

RMSprop: Adjusts learning rates using moving averages

Example (Adam):

optimizer = tf.keras.optimizers.Adam(learning_rate=0.01)

23. What is feature scaling? How does it help in Machine Learning?


Feature scaling standardizes or normalizes data to bring features to a common scale, improving model performance and convergence speed.

24. How do we perform scaling in Python?
